```python
import pandas as pd 
import matplotlib as plt
tabela = pd.read_excel('dataset.xlsx')
display(tabela)
```

```python
display(tabela.info())
```

```python
tabela = tabela.drop(tabela[tabela['Situação'] == 'APOSENTADO - 02'].index)
display(tabela.info())
```

```python
tabela = tabela[tabela['Situação'] != 'APOSENTADO - 02']
display(tabela.info())
```

```python
tabela = tabela[tabela['Cargo'] != 'PROFESSOR ENS BASICO TECN TECNOLOGICO (CMEBT) - 707001']
display(tabela.info())
```

```python
contagemCampus = tabela['Campus Exercício'].value_counts().reset_index()
contagemCampus.columns = ['Campus', 'Quantidade de Servidores']
pd.set_option('display.max_rows', None)
display(contagemCampus)
```

```python
contagemCargo = tabela['Cargo'].value_counts().reset_index()
pd.set_option('display.max_rows', None)
print(contagemCargo)
```

```python
#professores = tabela[tabela['Cargo'] == 'PROFESSOR ENS BASICO TECN TECNOLOGICO (CMEBT) - 707001']
#funcionarios = tabela[tabela['Cargo'] != 'PROF ENS BAS TEC TECNOLOGICO-SUBSTITUTO (CMEBT) - 707002']

#setores_professores = professores['Setor de Lotação']
#campus_professores = professores['Campus de Lotação']

#import pandas as pd
#import matplotlib.pyplot as plt


professores = tabela[tabela['Cargo'] == 'PROFESSOR ENS BASICO TECN TECNOLOGICO (CMEBT) - 707001']

setores_professores = professores['Campus Exercício']

porcentagens = setores_professores.value_counts(normalize=True) * 100

plt.figure(figsize=(12, 12))
ax = porcentagens.plot(kind='bar', color='darkgreen')

'''for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')'''

plt.title('Distribuição de Setores de Lotação dos Professores')
plt.xlabel('Setor de Lotação')
#plt.ylabel('Porcentagem (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
